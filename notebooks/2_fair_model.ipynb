# 2. Defining a FAIR brain network model

Sessions 2.1 · 2.2

Leon Stefanovski (Charité University Medicine Berlin)  
Marius Pille (Berlin Institute of Health at Charité University
Medicine)  
Leon Martin (Berlin Institute of Health at Charité University Medicine)

In this chapter you will use the TVB Ontology (TVB-O) to define a neural
mass model in a FAIR-compliant way, simulate a single node, and then
connect multiple nodes into a whole-brain network driven by structural
connectivity.

## Sessions

### [2.1 Mean field models](02a_mean_field_models.qmd)

Load a neural mass model from the TVB-O knowledge base. Inspect
parameters and state variables. Simulate a single uncoupled node and
visualise its dynamics.

### [2.2 Coupled networks](02b_coupled_networks.qmd)

Connect multiple neural masses via a structural connectome. Understand
how coupling functions and long-range delays shape collective network
activity.
