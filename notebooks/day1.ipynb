# Day 1

12 May 2026

An introduction to the mathematical and conceptual framework of
whole-brain network modelling using The Virtual Brain. Lectures
alternate with hands-on notebook sessions covering four chapters.

## Chapters

### [1. Introduction to TVB](1_introduction.qmd)

Basic ideas, history, mathematical framework, and applications of TVB.

### [2. Defining a FAIR brain network model](2_fair_model.qmd)

Define FAIR-compliant models with TVB-O, simulate single nodes, and
build whole-brain networks with structural connectivity.

-   [2.1 Mean field models](02a_mean_field_models.qmd)
-   [2.2 Coupled networks](02b_coupled_networks.qmd)

### [3. Dynamical perspective](3_dynamics.qmd)

Identify bifurcation points and understand how model parameters control
transitions between dynamical regimes.

-   [3.1 Bifurcation analysis](03_bifurcation_analysis.qmd)

### [4. Parameter exploration and optimisation](4_optimization.qmd)

Fit personalised models to empirical data using TVB-Optim.

-   [4.1 Alzheimer’s disease](04a_alzheimer_usecase.qmd)
-   [4.2 Parameter fitting](04b_parameter_fitting.qmd)

### [5. Stimulating the brain](5_stimulation.qmd)

Simulate visual and TMS-evoked potentials with the Jansen–Rit model.

-   [5.1 Stimulation](05_stimulation.qmd)

## What you need

-   Basic Python familiarity
-   An [EBRAINS](https://ebrains.eu) account to run the notebooks
    online, or a local Python environment set up via `uv sync`
