# 3. Dynamical perspective

Session 3.1 · Bifurcation analysis

Leon Stefanovski (Charité University Medicine Berlin)  
Marius Pille (Berlin Institute of Health at Charité University
Medicine)  
Leon Martin (Berlin Institute of Health at Charité University Medicine)

Neural mass models are dynamical systems. Before fitting them to data it
is essential to understand their bifurcation structure: which parameter
regions produce oscillations, fixed points, or multistability. This
chapter gives you the tools to map that landscape.

## Session

### [3 · Bifurcation analysis](03_bifurcation_analysis.qmd)

Use numerical continuation and phase-plane analysis to identify
bifurcation points in a neural mass model. Learn to interpret the
results and choose biologically plausible operating points for
whole-brain simulations.
