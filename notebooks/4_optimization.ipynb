# 4. Parameter exploration and inference

Sessions 4.1 · 4.2

Marius Pille (Berlin Institute of Health at Charité University
Medicine)  
Leon Martin (Berlin Institute of Health at Charité University
Medicine)  
Leon Stefanovski (Charité University Medicine Berlin)

In this chapter you will fit personalized whole-brain models to
empirical data using TVB-Optim’s gradient-based optimisation. Two
complementary approaches are covered: a disease use case with biomedical
data integration, and a systematic comparison of homogeneous versus
heterogeneous parameter strategies.

## Sessions

### [4a · Alzheimer’s disease use case](04a_alzheimer_usecase.qmd)

Build a personalised model using patient-specific connectivity and tau
deposition maps. Reproduce empirical BOLD signals and interpret the
result in the context of network degeneration.

### [4b · Parameter fitting](04b_parameter_fitting.qmd)

Compare homogeneous and heterogeneous parameter fitting strategies. Use
TVB-Optim to optimise a frequency gradient (or another map from TVBase)
and assess the impact on model fit quality.
