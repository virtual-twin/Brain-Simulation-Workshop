## About

Track 4 workshop of the **FAIR Brain Data Science Bootcamp**, 12-13 May
2026, Charité University Medicine Berlin and Berlin Institute of Health.

In this workshop, participants learn core concepts and methods for
personalized whole-brain network modeling using The Virtual Brain (TVB).
The workflow combines individualized model definition, simulation, and
analysis with ontology-driven model specification
([TVB-O](https://virtual-twin.github.io/tvbo)) and gradient-based
optimization tooling
([TVB-Optim](https://virtual-twin.github.io/tvboptim)) (Martin et al.
2025; Pille et al. 2025).

**Network structure**

Brain regions, tractography, surface geometry, and anatomical
constraints.

**Model dynamics**

Neural mass equations, coupling functions, integration, and
observations.

**Optimization loop**

Simulation targets, differentiable workflows, and parameter fitting.

## Getting Started

The Python environment is managed with [uv](https://docs.astral.sh/uv/):

``` bash
pip install uv
```

Tools used:

``` bash
uv pip install tvb
uv pip install tvbo
uv pip install tvboptim
```

The full source code lives at
<https://github.com/virtual-twin/Brain-Simulation-Workshop>.
